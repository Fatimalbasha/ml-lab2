## **Load the dataset**

```python
# Download the SMS Spam Collection dataset from Kaggle
import kagglehub
import os
import pandas as pd

path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")

print(path)
```

```python
csv_path = os.path.join(path, "spam.csv")
df = pd.read_csv(csv_path, encoding="latin-1")
```

## **Display the shape**

```python
df.shape
```

The dataset contains 5,572 rows (SMS messages) and 5 columns. Each row represents one SMS message

## **Preview the first few rows**

```python
df.head()
```

The first few rows show that column v1 contains the class labels (spam or ham), while column v2 contains the SMS message text. The remaining columns contain missing values

## **Check column names and data types**

```python
df.info()
```

The dataset consists of 5,572 entries. Columns v1 and v2 have no missing values and are of type object. The columns Unnamed: 2, Unnamed: 3, and Unnamed: 4 contain many missing values and are not relevant for the classification task

## **Extra Codes**
After inspecting the dataset, additional preprocessing steps will be applied to improve the data quality

```python
df = df.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
```

```python
df = df.rename(columns={"v1":"label", "v2":"text"})
```
